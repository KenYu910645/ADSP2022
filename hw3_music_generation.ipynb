```python
import math        #import needed modules
import pyaudio     #sudo apt-get install python-pyaudio
import numpy as np
from scipy.io import wavfile
PyAudio = pyaudio.PyAudio     #initialize pyaudio

# SCORE = [1,1,5,5,6,6,5]
# BEAT  = [1,1,1,1,1,1,2]

SCORE = [1,1,5,5,6,6,5 ,4,4,3,3,2,2,1 ,5,5,4,4,3,3,2 ,5,5,4,4,3,3,2, 1,1,5,5,6,6,5, 4,4,3,3,2,2,1]
BEAT  = [1,1,1,1,1,1,2 ,1,1,1,1,1,1,2 ,1,1,1,1,1,1,2 ,1,1,1,1,1,1,2, 1,1,1,1,1,1,2, 1,1,1,1,1,1,2]
NAME = "twinkle"

FREQ_MAP = {1:0, 2:2, 3:4, 4:5, 5:7, 6:9, 7:11, 8:12}
#See https://en.wikipedia.org/wiki/Bit_rate#Audio

BPM = 120
BITRATE = 44100 # 16000 # 44100 # 44100 # 16000     #number of frames per second/frameset.
BASE_FREQ = 261.63 # 261.63 # Hz, waves per second, 261.63=C4-note.
FILL_PERCENT = 0.8
TAU = 600 # for expotential decay 
SLOPE  = 1.0 # for linear decay 

N_FRAME_NOTE = int(BITRATE * (60 / BPM) * FILL_PERCENT)
N_FRAME_REST = int(BITRATE *  60 / BPM) - N_FRAME_NOTE
print(f"N_FRAME_NOTE = {N_FRAME_NOTE}")
print(f"N_FRAME_REST = {N_FRAME_REST}")

audio_num = []
for s, b in zip(SCORE, BEAT):
    feq = BASE_FREQ * (2**(FREQ_MAP[s]/12))
    #generating waves
    # Note

    for x in range(N_FRAME_NOTE*b + N_FRAME_REST*(b-1)):
        note = math.sin(x/((BITRATE/feq)/math.pi))
        # note = note * np.exp(-x/TAU) # exponential decay
        note = note * (1-SLOPE*(x/BITRATE))
        audio_num.append(math.sin(note/((BITRATE/feq)/math.pi)))
    # Rest
    for x in range(N_FRAME_REST): 
        audio_num.append(0)

# Save auiod to wav 
wavfile.write(f'{NAME}.wav', BITRATE, np.array(audio_num).astype(np.float32))

audio_str = "".join([chr(int(i*127+128)) for i in audio_num])
audio_str = bytes(audio_str, 'utf-8')

# Play audio 
p = PyAudio()
stream = p.open(format = p.get_format_from_width(1),
                channels = 1, 
                rate = BITRATE,
                frames_per_buffer=44,    
                output = True)

stream.write(audio_str)
stream.stop_stream()
stream.close()
p.terminate()
```
